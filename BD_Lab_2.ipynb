```python
# Importign necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics.pairwise import cosine_similarity
# Loading the dataset
df = fetch_20newsgroups(subset='all', shuffle=True)
doc = df.data
```

```python
# Creating a tfidf vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(doc)
```

```python
# Function to calculate cosine similarity between two tfidf vectors
def cosine_similarity_matrix(tfidf_matrix):
    return cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# Function to perform document similarity search
def document_similarity_search(query_doc):
    query_vector = vectorizer.transform([query_doc])
    similarity_scores = cosine_similarity(query_vector, tfidf_matrix)[0]
    most_similar_indices = similarity_scores.argsort()[::-1]

    # Getting the top 5 most similar documents
    sim_docs1 = [doc[index] for index in most_similar_indices[:5]]
    return sim_docs1
```

```python
# Executing the code
query_document = "baseball"
sim_docs = document_similarity_search(query_document)
print("most similar documents:\n")
for i, document in enumerate(sim_docs):
    print(f"Document {i+1}:\n{document}\n")
```
